```python
import numpy as np

def radial_component_S(E0, k, omega, mu):
    E0 = np.array(E0)
    S_radial = (k / (2 * omega * mu)) * np.linalg.norm(np.cross(E0, np.conj(E0)))
    return S_radial

#ejemplo
E0 = [1, 0, 0]  
k = 2 * np.pi / 1.0  
omega = 2 * np.pi * 1.0  
mu = 4 * np.pi * 1e-7  

S_radial = radial_component_S(E0, k, omega, mu)
print(f"Componente radial del vector de Poynting: {S_radial}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_poynting_radial(E0, k, omega, mu):
  
    theta = np.linspace(0, 2 * np.pi, 100)
    S_radial = radial_component_S(E0, k, omega, mu)
    r2S = S_radial * np.ones_like(theta)
    
    fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
    ax.plot(theta, r2S)
    ax.set_title(r'$r^2 \mathbf{S}(r) \cdot \hat{r}$ en el límite $r \to \infty$')
    plt.show()

# Ejemplo de uso:
plot_poynting_radial(E0, k, omega, mu)
```
